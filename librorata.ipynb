# LIBRO RATA EXCEPT - ELSE - FINALLY

El codigo se executa e encontra una excepcion en el try, en el caso  'spam' es una string con 4 caracteres, y el índice 99 enton 99 no es válido esto gera un error de índice.
Salta al except y lo executa con print.
Executa finally - El finally es opcional y siempre se ejecuta.
Se la excecion fue percorrida executa el resto del programa normalmente.

```python
sep = '-' * 45 + '\n'
print(sep + 'EXCEPTION RAISED AND CAUGHT')
try:
    x = 'spam'[99]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

En este caso except no se ejecuta pues no se produce ningum error dentro del try enton luego ejecuta finally y la ejecucion del programa.

```python
sep = '-' * 45 + '\n'
print(sep + 'NO EXCEPTION RAISED')
try:
    x = 'spam'[3]
except IndexError:
      print('except run')
finally:
      print('finally run')
print('after run')
```

En este, como no se produce ninguna excepción dentro del try,  except no se ejecuta. En su lugar, se ejecuta else. Por eso se imprime el contenido de else y luego el de finally.

```python
sep = '-' * 45 + '\n'
print(sep + 'NO EXCEPTION RAISED, WITH ELSE')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
else:
    print('else run')
finally:
    print('finally run')
    print('after run')
```

El except IndexError no captura esse erro, porque es otro tipo de excecion, se pongo ZeroDivisionError si que ejecuta.
Como el error no fue capturado, ejecuta o finally igual, despues
interrompe el programa y ejecuta la mensagem de error.
Por isso a linha print('after run') no se ejecuta. El codigo se ejecuta antes de ponder a excepcion.

```python
sep = '-' * 45 + '\n'
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT')
try:
    x = 1 / 0
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```
