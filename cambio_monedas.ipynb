# Ejemplo de cambio con monedas
calcular cómo dar cambio usando la menor cantidad de monedas posible.

# Definir las monedas y sus valores
Vamos a crear listas con los nombres y valores de las monedas.

```python
nombres_monedas = ['Nickels', 'Pennies', 'Dimes', 'Quarters']
valores_monedas = [5, 1, 10, 25]
print('Nombres:', nombres_monedas)
print('Valores:', valores_monedas)
```

# Crear un diccionario de monedas
Unimos las dos listas anteriores en un diccionario.

```python
billetera = dict(zip(nombres_monedas, valores_monedas))
print('Billetera:', billetera)
```

# Ordenar los valores de menor a mayor

```python
valores_ordenados = sorted(billetera.values())
print('Valores ordenados:', valores_ordenados)
```

# Calcular el cambio usando la menor cantidad de monedas

```python
cantidad_cambio = 56  # cantidad total a devolver
lista_cambio = []  # lista para guardar monedas usadas
i = len(valores_ordenados) - 1  # empezar por la moneda más grande
while cantidad_cambio > 0 and i >= 0:
    if cantidad_cambio >= valores_ordenados[i]:
        cantidad_cambio -= valores_ordenados[i]
        lista_cambio.append(valores_ordenados[i])
    else:
        i -= 1
print('Cambio en lista:', lista_cambio)
```

# Contar cuántas monedas de cada tipo usamos

```python
contador_cambio = dict.fromkeys(nombres_monedas, 0)  # inicializar en 0
for moneda in lista_cambio:
    for nombre in billetera:
        if billetera[nombre] == moneda:
            contador_cambio[nombre] += 1
print('Cambio en diccionario:', contador_cambio)
```
