<h1>OBJETOS INMUTABLES</h1>

Python almacena el primer valor que se le da a la variable a, en este caso es 3,
como b = a entonces b = 3, luego se le asigna a la variable a el valor spam, pero b ya 
tiene almacenado el valor 3, por lo tanto hasta que no se le asigne a b el nuevo 
valor de a, al llamar la funcion print (b) python seguira printando que b = 3. 

```python
a = 3
b = a               # b es 3
print("b es ", b)

a = 3
b = a
a = 'spam'         # b es 3
print("b es ", b)
```

a es un objeto integer, un objeto integer es inmutable, cuando 'se le da un nuevo valor' a la variable a, en realidad lo que se esta haciendo es creando una nueva variable con otro valor.

```python
a = 3
b = a
a = a + 2              # b es 3
print("b es ", b)
```

<h1>OBJETOS MUTABLES</h1>

En este caso, una lista si es un objeto mutable, se cambia el primer valor de L y por eso cuando se llama la funcion print(M), como M = L, tambien se cambia el primer valor de M.

```python

L = [2, 3, 4]	    # A mutable object
M = L			    # Make a reference to the same object
L[0] = 24 		    # An in-place change

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]
```

En este caso, M = L[:], esto significa que M != L, sino que es una copia de lo que es el primer valor que se le dio a L. Al cambiar el valor de L, no varia el valor de M.

```python
L = [2, 3, 4]
M = L[:]		    # Make a copy of L (or list(L), copy.copy(L), etc.)
L[0] = 24
print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M is not changed
```

## NESTED SCOPES


Cuando se usa = se le esta asignando valor a una variable y cuando se usa == se esta comparando si dos variables son iguales.

```python
L = [1, 2, 3]
M = L 			    # M and L reference the same object

L == M 			    # True: Same values => operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)

```

M y L tienen igual valor pero son objetos distintos, por eso is es falso pero == es verdadero 

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M and L reference different objects

print(
    "L == M",
    L == M 			# True: Same values
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```
